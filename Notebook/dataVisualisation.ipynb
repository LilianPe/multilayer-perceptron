# Corelation matrice
Allow to visualize relation between parameters

```python
numeric_data = data.select_dtypes(include=['float', 'int'])
corr_matrix = numeric_data.corr()
plt.figure(figsize=(22, 18))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrice de Corr√©lation')
plt.show()
```

La formule de la corr√©lation de Pearson est donn√©e par : $r = \frac{\sum_{i=1}^n (x_i - \overline{x})(y_i - \overline{y})}{\sqrt{\sum_{i=1}^n (x_i - \overline{x})^2} \sqrt{\sum_{i=1}^n (y_i - \overline{y})^2}}$
